### Mengambil data CSV

```python
#CSV

import csv
import urllib.request

def extract_csv(url, filename):
    #Download data dari URL
    urllib.request.urlretrieve (url, filename)

    #Membaca data CSV
    with open(filename, "r") as f:
        reader = csv.DictReader(f)
        data = [row for row in reader]
    return data

#Panggil fungsi
url = "https://hai.kemenkeu.go.id/file.php/3821WXSJDDDZAJ3820643058F58/FORMAT_237_SAKTI-1-1-.csv"
filename = "FORMAT_237_SAKTI-1-1-.csv"
data_sakti = extract_csv(url, filename)

data_sakti
```

### Mengambil data JSON

```python
#JSON

import json
import urllib.request

def extract_from_json(url, filename):
    #DownLoad data dari URL
    urllib.request.urlretrieve (url, filename)

    # Membaca data json
    with open(filename, "r") as f:
        reader = json.load(f)
        data = [row for row in reader]
    return data

# Panggil fungsi
url = "https://jsonplaceholder.typicode.com/posts"
filename = "posts.json"
data_posts = extract_from_json(url, filename)

data_posts
```

### Mengambil data dari API

```python
#API

import requests

def extract_api(url):
    # Mengambil data dari API
    response = requests.get(url)
    response.raise_for_status()

    #Convert data ke json
    data = response.json()
    return data

# Panggil fungsi
url = "https://api.spacexdata.com/v4/launches/latest"
data_spacex = extract_api(url)

data_spacex
```

### Mengambil data dari SITUS WEB

```python
#WEB
#Ambil isi dari https://ameera.republika.co.id/health

from bs4 import BeautifulSoup
import pandas as pd
import requests

page = requests.get("https://ameera.republika.co.id/health")
soup = BeautifulSoup(page.content, 'html.parser')

list_content_item = soup.find("div", {"class": "container-lit"})
list_result_item = []

for each in list_content_item.findAll("li"):
    data = dict()
    data['title'] = each.text
    data['link'] = each.a.get('href')
    list_result_item.append(data)

pd.DataFrame(list_result_item)
```

```python
#WEB
#Ambil table di www.pokemondb.net

from bs4 import BeautifulSoup
import pandas as pd
import requests

page = requests.get("https://pokemondb.net/pokedex/all")
soup = BeautifulSoup (page.content, 'html.parser')

table = soup.find("table", {"id":"pokedex"})
rows = table.findAll("tr")
row_data = []

for row in rows:
    cell_data = []

    if row.contents[1].get_text() == "301":
        break
    for item in row.findAll(["th", "td"]):
        cell_data.append(item.get_text())
    row_data.append(cell_data)

df_pokemon = pd.DataFrame(row_data)
df_pokemon.columns= df_pokemon.loc[0]
df_pokemon = df_pokemon [1:]
df_pokemon.head()
```

### Mengambil data Pandas Local

```python
#Pandas Local

import pandas as pd
import urllib.request

pd.options.display.max_columns = None

def extract_json_from_local_with_pandas (url, filename):
    #DownLoad data dari URL
    urllib.request.urlretrieve(url, filename)
    
    #Membaca data json
    df=pd.read_json(filename)
    return df

# Panggil fungsi
url = "https://jsonplaceholder.typicode.com/posts"
filename = "posts.json"
df_posts = extract_json_from_local_with_pandas(url, filename)

print(df_posts)
```
